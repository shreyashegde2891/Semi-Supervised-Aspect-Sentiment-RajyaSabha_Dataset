```python
def extract_string(text):
    split = text.split(';')
    question = split[0].replace('(a) ','')
    return question
```

```python
import nltk
nltk.download('punkt')
import re
def extract_primary_answer(answer):
    pattern = re.compile("[A-Z]")
    sentence_list = nltk.sent_tokenize(answer)
    primary_answer = []
    start = -1
    for item in sentence_list:
        if 'a)' in item:
            primary_answer.append(item)
    
    
    for i in reversed(primary_answer):
        answer_text = i
        #print("extracted answer text:")
       # print(answer_text)
        start = answer_text.index('a)')
        match = pattern.search(answer_text, start + 1)
        if match:
            start = match.start()
            return(answer_text[start:])
        
    
```
